```python
# Import Lib's 
import pandas as pd
from pandas.io import gbq
```

```python
# Insert the BigQuery Project ID
projectid = "capstone-196102"
```

```python
## Perform the query and save it to a pandas dataframe object 
## Need to FLATTEN two variables because they are nested
## Where "Capstone_Dataset" is the destination dataset and "blocks_copy" is the destination table

# Block level
block = gbq.read_gbq('SELECT * FROM (FLATTEN(FLATTEN(Capstone_Dataset.blocks_copy,transactions.inputs.input_pubkey_base58_error), transactions.outputs.output_satoshis)) LIMIT 25', projectid)

```

```python
block.head()
```

```python
# Transaction level
transaction = gbq.read_gbq('SELECT * FROM (FLATTEN(FLATTEN(Capstone_Dataset.transactions_copy, inputs.input_pubkey_base58_error), outputs.output_satoshis)) LIMIT 100', projectid)
```

```python
transaction.head()
```

```python
## Out put each query to a csv file 

# Block level
block.to_csv('block.csv', header=True)

# Transaction level
transaction.to_csv('transaction.csv', header=True)
```
